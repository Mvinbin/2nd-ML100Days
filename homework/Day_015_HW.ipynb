```r
library(ggplot2)
library(ggcorrplot)
library(GGally)
library(corrplot)

#請建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
data1=matrix(runif(10*10,-1,1),10,10)
colnames(data1)=c('X1','X2','X3','X4','X5','X6','X7','X8','X9','X10')
dat.corr=cor(data1)
plot1=ggcorrplot(dat.corr)
plot1

plot1.1=corrplot.mixed(dat.corr,lower.col = 'black',upper='col')
plot1.1
#請 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
data2=matrix(runif(1000*3,-1,1),1000,3)
colnames(data2)=c('X1','X2','X3')
data2=as.data.frame(data2)
plot2=ggpairs(data2,upper=list(continuous='points'),
        lower=list(continuous='density'),
        diag = list(continous='smooth'))
plot2

data.2.1=data2
data.2.1=data.frame(data2,type=as.factor(c(rep('type1',500),rep('type2',500))))
plot2.1=ggpairs(data.2.1,upper=list(continuous='points'),
                lower=list(continuous='density'),
                mapping=ggplot2::aes(col=type,alpha = 0.02),
                columns = c('X1','X2','X3'))
plot2.1

#請 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
data3=matrix(rnorm(1000*3),1000,3)
colnames(data3)=c('X1','X2','X3')
data3=as.data.frame(data2)
plot3=ggpairs(data2,upper=list(continuous='points'),
              lower=list(continuous='density'),
              diag = list(continous='smooth'))
plot3
```
