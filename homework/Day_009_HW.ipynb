```r
y=rnorm(1000,5,1)
x=-y+rnorm(1000,0,0.6)
dat=data.frame(x=x,y=y)
library(ggplot2)
plot=ggplot(data=dat,aes(x=x,y=y))+geom_point(color='skyblue4')+ggtitle('y versus x')+
  theme(plot.title =element_text(hjust = 0.5))+
  annotate('text', x=-3, y=8,label=paste('correlation=',round(cor(x,y),4)),size=5,color='blue4')
plot
```
