```r
library(data.table)
library(ggplot2)
library(gridExtra)
dat=fread('C:\\Users\\coco40725\\Documents\\GitHub\\2nd-ML100Days\\data\\application_train.csv')
house=(dat$HOUSETYPE_MODE)

## "  " imply this person doesn't have any house
house[which(house=='')]='houseless'
credict=(dat$AMT_CREDIT)

data.plot=data.frame(credict=credict,house.mode=house)

plot=ggplot(data=data.plot,aes(x=credict,fill=house.mode))+geom_histogram()+
  facet_wrap(~house.mode,scales = 'free')+
  ggtitle('Relationship between house mode and credict ')+
  theme(plot.title = element_text(hjust = 0.5))
plot
```
