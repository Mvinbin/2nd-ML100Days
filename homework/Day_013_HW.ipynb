```r
library(data.table)
library(ggplot2)
dat=fread('C:\\Users\\coco40725\\Documents\\GitHub\\2nd-ML100Days\\data\\application_train.csv')
income=dat$AMT_INCOME_TOTAL
##remove outlier
income=income[-which(income==max(income))]
group.income=cut(income,breaks=c(10^3,5*10^3,
                                 10^4,5*10^4,
                                 10^5,5*10^5,
                                 10^6,5*10^6,
                                 10^7,5*10^7,
                                 10^8))

group.income.dat=as.data.frame(table(group.income))

plot=ggplot()+
  geom_bar(stat = 'identity',data=group.income.dat,aes(x=group.income,y=Freq,fill=group.income))+
  ggtitle('income discrete')+theme(plot.title = element_text(hjust = 0.5))+
  theme(axis.text =element_text(angle = 90) )
table(group.income)
plot
```
