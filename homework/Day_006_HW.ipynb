## 尋找哪個column可能出現outlier

```r
data=read.csv('C:\\Users\\coco40725\\Documents\\GitHub\\2nd-ML100Days\\data\\application_train.csv')
library("dplyr")
data.cont=select_if(data, is.numeric)

##去掉categorial data
## 只挑選連續型資料
col.len=c()
for (i in 1:ncol(data.cont)) {
  col.len[i]=length(unique(data.cont[,i]))
}
remove.loc=which(col.len<=10)
cont.data.seclect=data.cont[,-remove.loc]

##67個欄
mean=max=min=median=c()
for (j in 1:ncol(cont.data.seclect)) {
  mean[j]=mean(cont.data.seclect[,j],na.rm = T)
  max[j]=max(cont.data.seclect[,j],na.rm = T)
  min[j]=min(cont.data.seclect[,j],na.rm = T)
  median[j]=median(cont.data.seclect[,j],na.rm = T)
}
all.table=data.frame(type=colnames(cont.data.seclect),min=min,mean=mean,median=median,max=max)
all.table

```

### 可疑欄位:  AMT_INCOME_TOTAL、 OBS_60_CNT_SOCIAL_CIRCLE、DAYS_EMPLOYED
## 畫ECDF 觀察

```r
##可疑欄位:  AMT_INCOME_TOTAL、 OBS_60_CNT_SOCIAL_CIRCLE、DAYS_EMPLOYED
##繪製 Empirical Cumulative Density Plot (ECDF)
suspect.data=all.table[c(3,9,63),]
suspect.data

library(ggplot2)
library(gridExtra)
plot1=ggplot(data=cont.data.seclect,aes(x=AMT_INCOME_TOTAL))+stat_ecdf(color='red',size =1)+
  ggtitle('AMT_INCOME_TOTAL ECDF')+theme(plot.title = element_text(hjust = 0.5))

plot2=ggplot(data=cont.data.seclect,aes(x=OBS_60_CNT_SOCIAL_CIRCLE))+stat_ecdf(color='blue4',size =1)+
  ggtitle('OBS_60_CNT_SOCIAL_CIRCLE ECDF')+theme(plot.title = element_text(hjust = 0.5))

plot3=ggplot(data=cont.data.seclect,aes(x=DAYS_EMPLOYED))+stat_ecdf(color='green3',size =1)+
  ggtitle('DAYS_EMPLOYED ECDF')+theme(plot.title = element_text(hjust = 0.5))

grid.arrange(plot1,plot2,plot3)

```

### ECDF出現"長"橫線

## 去掉outlier的boxplot

```r
##去掉outlier的boxplot
remove1.loc=which(cont.data.seclect$AMT_INCOME_TOTAL==max(cont.data.seclect$AMT_INCOME_TOTAL))
remove2.loc=which(cont.data.seclect$OBS_60_CNT_SOCIAL_CIRCLE==max(cont.data.seclect$OBS_60_CNT_SOCIAL_CIRCLE,na.rm=T))
remove3.loc=which(cont.data.seclect$DAYS_EMPLOYED==max(cont.data.seclect$DAYS_EMPLOYED))

r.bplot1=ggplot(data=cont.data.seclect[-remove1.loc,],aes(y=AMT_INCOME_TOTAL))+geom_boxplot(color='red')+
  ggtitle('INCOME without outlier')+theme(plot.title = element_text(hjust = 0.5))

r.bplot2=ggplot(data=cont.data.seclect[-remove2.loc,],aes(y=OBS_60_CNT_SOCIAL_CIRCLE))+geom_boxplot(color='blue4')+
  ggtitle('OBS_60 without outlier')+theme(plot.title = element_text(hjust = 0.5))

r.bplot3=ggplot(data=cont.data.seclect[-remove3.loc,],aes(y=DAYS_EMPLOYED))+geom_boxplot(color='green3')+
  ggtitle('DAYS_EMPLOYED without outlier')+theme(plot.title = element_text(hjust = 0.5))

grid.arrange(r.bplot1,r.bplot2,r.bplot3,ncol=3)
```

### 去掉outlier後，boxplot的方形結構會較為明顯
